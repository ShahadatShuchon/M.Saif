# Month Extraction

```python
luzerne_gps_rev['Date'] = pd.to_datetime(luzerne_gps_rev.Date, format='%Y-%m-%d %H:%M:%S')
luzerne_gps_rev['Month'] = luzerne_gps_rev['Date'].dt.month
luzerne_gps_rev['Month'] = luzerne_gps_rev['Date'].dt.strftime('%B')
luzerne_gps_rev.head(2)
```

# season inclution

```python
def label_race (row):
   if row['Month'] == 'May' :
      return 'Spring'
   if row['Month'] == 'June' :
      return 'Spring'
   if row['Month'] == 'July' :
      return 'Spring'
   if row['Month'] == 'August':
      return 'Fall'
   if row['Month']  == 'September':
      return 'Fall'
   if row['Month'] == 'October':
      return 'Fall'
   if row['Month'] == 'November':
      return 'Fall'
   if row['Month'] == 'December':
      return 'Fall'
```

```python
luzerne_gps_rev['Season'] = luzerne_gps_rev.apply (lambda row: label_race(row), axis=1)
luzerne_gps_rev.head(2)
```
